```python
# Eliminar la columna CustomerID
data = data.drop(columns=['CustomerID'])
```

```python
# Convertir la columna Gender a valores numéricos (por ejemplo, 0 para Male y 1 para Female)
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})
```

```python
from sklearn.preprocessing import StandardScaler

# Escalar las variables numéricas
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Lista para almacenar las inercias
inertias = []

# Probar diferentes valores de k (de 1 a 10 por ejemplo)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertias.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para Determinar k')
plt.show()
```

```python
# Entrenar el modelo K-means con k=4 (ajusta k según el resultado del gráfico del codo)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data_scaled)

# Obtener los clusters predichos
clusters = kmeans.predict(data_scaled)

# Agregar los clusters como una nueva columna en el DataFrame original para análisis
data['Cluster'] = clusters
```

```python
from sklearn.decomposition import PCA

# Aplicar PCA y reducir a 2 componentes principales (o ajustarlo según tu criterio)
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Crear un DataFrame con las componentes principales para facilitar el análisis
data_pca_df = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])

# Verificar la varianza explicada por cada componente
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
```

```python
# Entrenar el modelo K-means con los datos reducidos
kmeans_pca = KMeans(n_clusters=4, random_state=42)
kmeans_pca.fit(data_pca)

# Agregar los clusters como una nueva columna en el DataFrame de componentes principales
data_pca_df['Cluster'] = kmeans_pca.labels_

# Visualizar los resultados en un gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(data_pca_df['PC1'], data_pca_df['PC2'], c=data_pca_df['Cluster'], cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clusters de Clientes (K-means con PCA)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Configuraciones de hiperparámetros a probar
params = {
    'n_clusters': [3, 4, 5],
    'init': ['k-means++', 'random'],
    'max_iter': [300, 500],
    'n_init': [10, 20],
    'random_state': 42
}

# Entrenar y comparar modelos con diferentes configuraciones
for n_clusters in params['n_clusters']:
    for init in params['init']:
        for max_iter in params['max_iter']:
            for n_init in params['n_init']:
                kmeans = KMeans(
                    n_clusters=n_clusters,
                    init=init,
                    max_iter=max_iter,
                    n_init=n_init,
                    random_state=params['random_state']
                )
                kmeans.fit(data_scaled)

                # Evaluar y mostrar el resultado de inercia (inertia) como métrica de desempeño
                print(f'n_clusters: {n_clusters}, init: {init}, max_iter: {max_iter}, n_init: {n_init}, inertia: {kmeans.inertia_}')
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Calcular el Coeficiente de Silhouette
silhouette_avg = silhouette_score(data_scaled, clusters)
print(f"Coeficiente de Silhouette: {silhouette_avg}")

# Calcular el Índice de Calinski-Harabasz
calinski_harabasz = calinski_harabasz_score(data_scaled, clusters)
print(f"Índice de Calinski-Harabasz: {calinski_harabasz}")
```
